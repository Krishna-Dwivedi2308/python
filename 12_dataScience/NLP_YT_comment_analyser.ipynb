```python
import pandas as pd
import numpy as np
```

```python
toxic_comments=[
    "you're so dumb","This is a trash","Worst video ever","Stop making content","You sound horrrible","Clickbait title","Can't believe people like this","Waste of time","Cringe Content","You're such a loser"
    ]
```

```python
supportive_comments=[
    "This helped me a lot","You're amazing","Best tutorial I've seen","Thanks for the content","Keep up the great work","So clear and helpful","Awesome explanation","I learned a lot","Great job","You're a genius"
    ]
```

```python
import random
data=[]
for i in range(50):
    data.append({"comment":random.choice(toxic_comments),"label":'toxic'})
    data.append({"comment":random.choice(supportive_comments),"label":'supportive'})
df=pd.DataFrame(data)
df.head()
```

```python
df.to_csv("comments.csv",index=False)
```
